# **Data Cleaning**

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

df = df.drop('Cabin', axis=1)

df['Sex'] = df['Sex'].map({'male':0, 'female':1})

df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)

df = df.drop(['PassengerId','Name','Ticket'], axis=1)

df.info()
df.isnull().sum()
```

```python
# Summary statistics for numeric columns
summary_stats = df.describe()
print(summary_stats)

# Summary statistics including categorical
full_summary = df.describe(include='all')
print(full_summary)

# Missing values check
missing_values = df.isnull().sum().sort_values(ascending=False)
print(missing_values)
```

```python
survival_by_class = df.groupby('Pclass')['Survived'].agg(['mean','count'])
print(survival_by_class)
```

```python
gender_class_survival = df.groupby(['Sex','Pclass'])['Survived'].mean().unstack()
print(gender_class_survival)
```

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[0,12,18,35,60,100],
                        labels=['Child','Teen','YoungAdult','Adult','Senior'])

age_survival = df.groupby('AgeGroup')['Survived'].mean()
print(age_survival)
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

family_survival = df.groupby('FamilySize')['Survived'].mean()
print(family_survival)
```

```python
pivot_table = pd.pivot_table(
    df,
    values='Survived',
    index='Sex',
    columns='Pclass',
    aggfunc='mean'
)

print(pivot_table)
```
